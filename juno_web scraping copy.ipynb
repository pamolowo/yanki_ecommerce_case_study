```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd

# Configure Selenium ChromeDriver options
options = Options()
options.add_argument('--headless')  # Run Chrome in headless mode
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.6533.72 Safari/537.36')

# Set up the Chrome driver service
chrome_driver_path = r'C:\Users\PAMELA\Downloads\chromedriver-win64\chromedriver.exe'  # Replace with the correct path to your chromedriver
service = Service(chrome_driver_path)

# Create a new Chrome browser instance
driver = webdriver.Chrome(service=service, options=options)

# List of pages to scrape
pages = [x for x in range(1, 51)]

product_names = []
current_prices = []
former_prices = []
discount_percs = []

for page in pages:
    # Open the website
    url = f'https://www.jumia.com.ng/catalog/?q=laptop&page={page}#catalog-listing'
    driver.get(url)

    # Wait for the page to load and the JavaScript to execute
    driver.implicitly_wait(10)

    # Get the page source
    html = driver.page_source

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    
    # Extract product names
    names = soup.find_all("h3", class_="name")
    for name in names:
        product_names.append(name.get_text(strip=True))

    # Extract current prices
    prices = soup.find_all("div", class_="prc")
    for price in prices:
        current_prices.append(price.get_text(strip=True))

    # Extract former prices
    old_prices = soup.find_all("div", class_="old")
    for old_price in old_prices:
        former_prices.append(old_price.get_text(strip=True))

    # Extract discount percentages
    discounts = soup.find_all("div", class_="bdg _dsct _sm")
    for discount in discounts:
        discount_percs.append(discount.get_text(strip=True))

# Once all the pages information are extracted, quit the driver
driver.quit()

# Create a DataFrame
data = {
    'title': product_names,
    'current_price': current_prices,
    'former_price': former_prices,
    'discount': discount_percs
}
laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
laptop_df = laptop_df.transpose()

# Display the DataFrame
# print(laptop_df)
```

```python
# for all the pages
pages = [x for x in range(1,51)]

product_names = []
current_prices = []
former_prices = []
discount_percs = []

for page in pages:
    url = f"https://www.jumia.com.ng/catalog/?q=laptops&page={page}#catalog-listing"

    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")

    # Product names
    names = soup.find_all("h3", class_ = "name")
    for name in names:
        product_names.append(name.string)

    # Current price
    prices = soup.find_all("div", class_ = "prc")
    for price in prices:
        current_prices.append(price.string)

    # former price
    prices = soup.find_all("div", class_ = "old")
    for price in prices:
        former_prices.append(price.string)

    # discount percentages
    discounts = soup.find_all("div", class_ = "bdg _dsct _sm")
    for discount in discounts:
        discount_percs.append(discount.string)

data = {
    'title': product_names, 
    'current_price': current_prices, 
    'former_price': former_prices, 
    'discount': discount_percs
    }
laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
laptop_df = laptop_df.transpose()
laptop_df
```
