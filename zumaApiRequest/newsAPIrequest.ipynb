```python
import requests
import json
import pandas as pd
```

```python


API_key = "c356dec563044b308bb1fc9c399e11fc"  # Replace with your actual NewsAPI key
endPoint = "https://newsapi.org/v2/everything"
headers = {
    "Content-Type": "application/json"
}

# Define the keywords to search for
keywords = ["Bitcoin", "Ethereum", "Litecoin", "BNB"]

# Function to fetch news articles
def fetch_news(api_key, endpoint, keywords):
    results = []
    for keyword in keywords:
        params = {
            'q': keyword,
            'apiKey': api_key,
            'language': 'en',
            'pageSize': 100  # Number of articles to retrieve per keyword
        }
        response = requests.get(endpoint, headers=headers, params=params)
        if response.status_code == 200:
            articles = response.json().get('articles', [])
            results.extend(articles)
        else:
            print(f"Failed to retrieve news for keyword: {keyword}, Status code: {response.status_code}")
    return results

# Fetch news articles for the given keywords
news_articles = fetch_news(API_key, endPoint, keywords)

# Print the titles of the fetched articles
if news_articles:
    for article in news_articles:
        print(f"Title: {article.get('title')}, URL: {article.get('url')}, Author:{article.get('author')}, Description:{article.get('description')}, Content:, 'URL', 'Published_Date'")
else:
    print("No articles were fetched.")

# Save the articles to a JSON file
with open('news_articles.json', 'w') as f:
    json.dump(news_articles, f, indent=4)

print("News articles fetched and saved successfully.")
```
